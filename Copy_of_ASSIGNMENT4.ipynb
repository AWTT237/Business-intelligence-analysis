```python
# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
# Task 8: Functions and Lambda

# Regular function to calculate the square of a number
def square(num):
    return num ** 2

# Test the regular function
print("Square using function:", square(5))

# Lambda function to calculate the square
square_lambda = lambda num: num ** 2

# Test the lambda function
print("Square using lambda:", square_lambda(5))
```

```python
# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Task 9: Iterators and Generators

# Create an iterator for the first 5 even numbers
even_numbers_list = [2, 4, 6, 8, 10]
even_iterator = iter(even_numbers_list)

print("Iterator output:")
for num in even_iterator:
    print(num)

# Create a generator for the first 5 even numbers
def even_generator():
    for i in range(2, 11, 2):
        yield i

print("\nGenerator output:")
for num in even_generator():
    print(num)
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
# Task 10: Map, Reduce, and Filter
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# Use map to square all numbers
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared numbers using map:", squared_numbers)

# Use reduce to find the product of all numbers
product = reduce(lambda x, y: x * y, numbers)
print("Product of all numbers using reduce:", product)

# Use filter to get only even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers using filter:", even_numbers)

```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
# Task 11: Object-Oriented Programming - Creating a Class

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    # Method to calculate area
    def area(self):
        return self.length * self.width

    # Method to calculate perimeter
    def perimeter(self):
        return 2 * (self.length + self.width)

# Create an instance of Rectangle
rect1 = Rectangle(5, 3)

# Test the methods
print("Area of rectangle:", rect1.area())
print("Perimeter of rectangle:", rect1.perimeter())
```

```python
# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
import pandas as pd

# Sample DataFrame
df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# 1. Find the average salary by department
avg_salary_by_dept = df_employees.groupby('Department')['Salary'].mean()
print("Average salary by department:\n", avg_salary_by_dept)

# 2. Get the names of employees with salary > 60000
high_earners = df_employees[df_employees['Salary'] > 60000]['Name'].tolist()
print("\nEmployees with salary > 60000:", high_earners)

# 3. Add a new column 'Bonus' which is 10% of salary
df_employees['Bonus'] = df_employees['Salary'] * 0.10
print("\nDataFrame with Bonus column:\n", df_employees)
```
